### **Outliers Values**

Memeriksa nilai ekstrem (Outliers) pada kolom yang bertipe data numerik, agar mendapat insight yang lebih akurat sehingga dapat memperkuat analisis dan model nantinya

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, '/ Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Outliers**

Outlier adalah data yang nilainya menyimpang jauh dari mayoritas data lain dalam suatu dataset. Dengan kata lain, outlier merupakan anomali atau pencilan yang tidak mengikuti pola umum. Kehadiran outlier bisa memengaruhi hasil analisis

**Memeriksa Persentase kemungkinan adanya outlier :**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

#Dataframe dari list hasil

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

#Tampilkan dataframe

display(results_df)
```

Berdasarkan output di atas, terlihat hasil yang menunjukkan adanya outlier pada beberapa kolom di dataset ini. Hal ini dapat menyebabkan hasil analisis menjadi kurang akurat apabila tidak ditangani terlebih dahulu.

**Penanganan Outlier:**

```python
columns_to_impute = ["price", "threads_num", "core_num", "ram_num", "memory_size", "display_size","resolutioin_height", "resolutioin_width", "warrenty"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

**Metode Penanganan :**

Metode penanganan outlier yang digunakan adalah **Capping**. Metode ini mengganti nilai outlier dengan nilai batas tertentu, sehingga nilai outlier tidak dihapus melainkan dibatasi supaya nilainya tidak terlalu ekstrem.

Metode ini dipilih karena data pada kolom-kolom tersebut banyak yang sangat berharga, dan dapat mengubah distribusi dataset secara drastis apabila kolom tersebut dihapus.

Dengan penggunaan metode **Capping**, metode ini akan menjaga supaya nilai yang ekstrem tadinya tidak melebihi batas yang wajar, dengan membatasi nilai-nilai yang ada supaya data tersebut masih dapat dipertahankan.


Periksa hasil penanganan outlier setelah menggunakan metode **Capping** :

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

#Dataframe dari list hasil

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

#Tampilkan dataframe

display(results_df)
```

Dengan demikian, nilai-nilai ektrem tadi menjadi dalam jangkauan nilai yang wajar, sehingga proses analisis dapat dilanjutkan ke tahap selanjutnya.

# **Data Construction**

Berdasarkan studi kasus dan tema yang diangkat, dataset ini memerlukan Data Construction berupa **Kolom Kategori Harga**. Kolom ini akan mengategorikan harga sebuah laptop, seperti kelas "Low-end", "Mid-range", dan "Premium" yang dikategorikan berdasarkan rentang harganya.

```python
low_price_max = df['price'].quantile(0.33)
medium_price_max = df['price'].quantile(0.66)

def categorize_price(price):
    if price <= low_price_max:
        return 'Low-end'
    elif price <= medium_price_max:
        return 'Mid-end'
    else:
        return 'High-end'

df['Kategori Harga'] = df['price'].apply(categorize_price)

# menampilkan beberapa baris dengan kolom baru
display(df[['price', 'Kategori Harga']].head())

print("\nDistribusi Kategori Harga:")
display(df['Kategori Harga'].value_counts())
```

Pada hasil data construction di atas, kita membuat kolom baru yaitu **Kategori_harga** yang mengelompokkan harga laptop menjadi tiga kategori (Low, Medium, High) berdasarkan rentang harga.

Dengan demikian, proses analisis untuk menentukan rekomendasi laptop berdasarkan harga sesuai tema yang diambil akan menjadi lebih relevan, karena memberikan data berupa kategori kelas laptop berdasarkan rentang harganya.

## **Heatmap**

```python
# Membuat heatmap korelasi khusus Price dan RAM
plt.figure(figsize=(8, 6))
corr_price_ram = df[['price', 'ram_num']].corr()

sns.heatmap(corr_price_ram, annot=True, cmap='YlGnBu', fmt='.2f', square=True, annot_kws={'size': 15})
plt.title('Korelasi antara Harga dan Kapasitas RAM')
plt.tight_layout()
plt.show()
```

Visualisasi di atas menjelaskan korelasi antara **Harga sebuah laptop** dengan **Kapasitas RAM**. Hal ini bertujuan untuk menampilkan apakah kapasitas RAM pada sebuah laptop memiliki pengaruh terhadap harga unit laptop tersebut.

### **Insight : Korelasi Harga laptop dan Kapasitas RAM**

- Nilai korelasi sekitar 0.71 menunjukkan adanya hubungan positif yang kuat. Artinya, Kapasitas RAM adalah salah satu variabel yang paling berpengaruh secara langsung dalam menentukan kenaikan harga sebuah laptop.
- Karena nilainya mendekati 1, kita bisa menyimpulkan bahwa setiap peningkatan kapasitas RAM (misal dari 8GB ke 16GB) hampir selalu diikuti dengan kenaikan harga yang signifikan di pasar.
